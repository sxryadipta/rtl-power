```python
import os

files = [f for f in os.listdir() if f.endswith(".v")]

data = []

for f in files:
    cmd = f'yosys -p "read_verilog {f}; synth; stat"'
    output = os.popen(cmd).read()

    lines = output.split("\n")

    cells = 0
    wires = 0

    for line in lines:
        if "Number of cells:" in line:
            cells = int(line.split()[-1])
        if "Number of wires:" in line:
            wires = int(line.split()[-1])

    power = cells * 0.5 + wires * 0.2

    data.append([f, cells, wires, power])
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=["file", "cells", "wires", "power"])
df.to_csv("dataset.csv", index=False)
df.head()
```
